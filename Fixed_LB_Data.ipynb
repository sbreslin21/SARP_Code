```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
```

```python
file_path = './Longbeach_data_fixed.csv'
```

```python
data=pd.read_csv(file_path)
```

```python
data
```

```python
data.drop(index=range(22, 30), inplace=True)  # Drop rows with index from 22 to 29 (inclusive of 22, exclusive of 30)
```

```python
data
```

```python
cleaned_data=data[['Flight','Jday','date', 'Latitude', 'Longitude', 'AltP_feet', 'CH4 (ppmv)','CO (ppbv)','Ethane (E)']]
```

```python
cleaned_data
```

```python
cleaned_data['Methane/CO (ppb)']=((cleaned_data['CH4 (ppmv)'])*1000)/cleaned_data['CO (ppbv)']
```

```python
cleaned_data
```

```python
cleaned_data['Ethane/CO (ppb)']=((cleaned_data['Ethane (E)'])/1000)/cleaned_data['CO (ppbv)']
```

```python
cleaned_data
```

```python
gases=cleaned_data[['Jday','CH4 (ppmv)','CO (ppbv)','Ethane (E)','Methane/CO (ppb)','Ethane/CO (ppb)']]
```

```python
gases['CH4 (ppbv)']=gases['CH4 (ppmv)']*1000
gases['Ethane (ppbv)']=gases['Ethane (E)']/1000
```

```python
gases
```

```python
converted_gases=gases[['Jday','CH4 (ppmv)','CO (ppbv)','Ethane (E)','CH4 (ppbv)','Ethane (ppbv)']]
```

```python
converted_gases
```

```python
summer_data=gases[(gases['Jday'] < 300)]
winter_data=gases[(gases['Jday'] > 300)]
```

```python
summer_data
```

```python
winter_data
```

```python
winter_data['Ethane (ppmv)']=winter_data['Ethane (ppbv)']/1000
```

```python
winter_data
```

```python
x = summer_data['CO (ppbv)']
y = summer_data['CH4 (ppbv)']

regr = linear_model.LinearRegression()

# Convert x and y data from a pandas series type to a numpy array type
regr_x = x.values
regr_y = y.values


# Reshaping array from 1D to 2D
regr_x = regr_x.reshape(len(regr_x),1)
regr_y = regr_y.reshape(len(regr_y),1)

# Input is ready to be passed as arguments to the following .fit function - a linear model is created
regr.fit(regr_x,regr_y)
print('Regression Intercept:  ', regr.intercept_)
regr_slope = (regr.predict(regr_x)[2] - regr.predict(regr_x)[1])/ (regr_x[2] - regr_x[1])
print('Regression Slope: ', regr_slope)

equation = 'y = ' + str(round(float(regr_slope),3)) + '*x + ' + str(round(float(regr.intercept_),3))
print('Regression Equation: ', equation)

r_squared = regr.score(regr_x, regr_y)
print('R^2 value: ', r_squared)

fig = plt.figure()
ax = fig.add_subplot()

plt.scatter(x, y)
plt.scatter(x, y, color='red', linestyle='--')
plt.plot(regr_x, regr.predict(regr_x), color='orange', linewidth=1.5) # Passing regr_x values into the model that we created
plt.title('Summer 2022 CH4 vs CO')    
plt.xlabel('CO (ppbv)')               
plt.ylabel('CH4 (ppbv)')                      
plt.rcParams["figure.figsize"] = (8,5)  
ax.text(135,1995, equation) # Add text
ax.text(135, 1993, f'R^2 = {round(r_squared, 3)}')  # Add R^2 value to the plot
```
